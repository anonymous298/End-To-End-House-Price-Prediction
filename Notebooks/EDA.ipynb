# Visualizing Data
## (Univariate Anaylsis)

### Visualizing the distribution of Price

```python
plt.figure(figsize=(15,7))
plt.subplot(1,3,1)
sns.histplot(data=df, x='price', kde=True, bins=30, palette='viridis')

plt.subplot(1,3,2)
sns.histplot(data=df, x='price', kde=True, hue='total_rooms', bins=30, palette='viridis')

plt.subplot(1,3,3)
sns.histplot(data=df, x='price', kde=True, hue='floors', bins=30, palette='viridis')
```

### Insights
- The House Prices are very high on 1935, 2000, 2020
- The prices are too low at year 1940 to 1943

### Finding distribution of price

```python
sns.violinplot(data=df, y='yr_built')
```

- Houses were mostly build in years ranges between 1940 to 2020

### Visualizing the counts of rooms in houses

```python
sns.countplot(data=df, x='total_rooms')
```

### Insights
- houses are more likely to have 5 total_rooms including bath rooms

### Visualizing the Houses with most floors

```python
sns.countplot(data=df, x='floors')
```

### Insights
- Houses are more likely to have 1 floor and also Most of the houses have 2 floors

### Visualzing Prices of house based on view, waterfront and condition

```python
plt.figure(figsize=(15,7))
plt.subplot(1,3,1)
sns.barplot(data=df, x='waterfront', y='price', hue='waterfront')

plt.subplot(1,3,2)
sns.barplot(data=df, x='view', y='price', hue='view')

plt.subplot(1,3,3)
sns.barplot(data=df, x='condition', y='price', hue='condition')
```

### Insights
- Houses with waterfront have more higher prices
- houses with 4 no view have more prices
- houses with 5/5 condition has ofcourse higher prices

```python
plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
sns.histplot(data=df, x='yr_built', kde=True)
```

- Peak year when most houses are build is 2000 to 2020

# (Multivariate - Analysis)

### Visualzing the pairplot

```python
sns.pairplot(df)
```

### Visualizing the correlations of each columns

```python
plt.figure(figsize=(20, 13))
sns.heatmap(df.corr(), annot=True)
```

### Insights
- Higher rooms are tend to increase prices
- Higher sqft tend to increase in price
- View is also a plus point in a house that why the price increases if view is solid

### Visualizing the prices based on year

```python
sns.lineplot(data=df, x='yr_built', y='price')
```

### Insights
- year 1935, 2000 and 2020+ has more higher prices of houses

### Visualizing the relationship between price and sqft_living

```python
plt.figure(figsize=(15,8))
plt.subplot(1,2,1)
sns.scatterplot(data=df, x='sqft_living', y='price')

plt.subplot(1,2,2)
sns.scatterplot(data=df, x='sqft_living', y='price', hue='floors')
```

### Insights 
- increase in sqft tend to increase the house prices
- Higher floors are not so increasing the price

### Visualizing the relation between price and sqft_basement

```python
sns.scatterplot(data=df, x='sqft_basement', y='price')
```

### Insights
- Basement is a plus point in a house but it is not directly correlated to increase in price.
